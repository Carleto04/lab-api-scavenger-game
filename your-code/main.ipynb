## Challenge 3: Hidden Cold Joke

Using Python, call Github API to find out the cold joke contained in the 24 secret files in the following repo:

https://github.com/ironhack-datalabs/scavenger

The filenames of the secret files contain .scavengerhunt and they are scattered in different directories of this repo. The secret files are named from .0001.scavengerhunt to .0024.scavengerhunt. They are scattered randomly throughout this repo. You need to search for these files by calling the Github API, not searching the local files on your computer.

#### 1. Find the secret files.

`````python
url = 'https://api.github.com/repos/ironhack-datalabs/scavenger/contents'
folders_scav_response = requests.get(url, auth=(username, token))
folders_scav_json = folders_scav_response.json()

# To print the JSON response in a pretty way
print(json.dumps(folders_scav_json, indent=6))
`````

`````python
list_of_folder = []

for i in range(1, len(folders_scav_json)):
    list_of_folder.append(folders_scav_json[i]['name'])
    
list_of_folder
`````

`````python
list_of_folders_contents = []

for i in list_of_folder:
    url = f'https://api.github.com/repos/ironhack-datalabs/scavenger/contents/{i}'
    temp = requests.get(url, auth=(username, token))
    jsons_content = temp.json()
    list_of_folders_contents.append(jsons_content)
    
list_of_folders_contents
`````

`````python
dictionary = {}
for i in list_of_folders_contents:
    for l in i:
        if 'hunt' in l['name']:
            dictionary[l['name']] = l['path']
`````

#### 2.  Sort the filenames ascendingly.

`````python
sorted_dict = dict(sorted(dictionary.items()))

sorted_dict
`````

`````python
sorted_dict.keys()
`````

#### 3. Read the content of each secret files into an array of strings.
Since the response is encoded, you will need to send the following information in the header of your request:
````python
headers = {'Accept': 'application/vnd.github.v3.raw'}
````

`````python
list_of_folders_contents2 = []

for i in sorted_dict.values():
    headers = {'Accept': 'application/vnd.github.v3.raw'}
    url = f'https://api.github.com/repos/ironhack-datalabs/scavenger/contents/{i}'
    temp2 = requests.get(url, auth=(username, token), headers = headers)
    text_content2 = temp2.text
    list_of_folders_contents2.append(text_content2)
`````

`````python
list_of_folders_contents2
`````

#### 4. Concatenate the strings in the array separating each two with a whitespace.

`````python
list_of_folders_contents2 = "".join(list_of_folders_contents2)
list_of_folders_contents2 = list_of_folders_contents2.replace('\n', ' ')
`````

#### 5. Print out the joke.
